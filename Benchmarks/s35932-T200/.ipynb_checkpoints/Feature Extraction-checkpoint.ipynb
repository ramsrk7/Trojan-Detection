```python
#import libraries
import progressbar as pb

#define progress timer class
class progress_timer:

    def __init__(self, n_iter, description="Something"):
        self.n_iter         = n_iter
        self.iter           = 0
        self.description    = description + ': '
        self.timer          = None
        self.initialize()

    def initialize(self):
        #initialize timer
        widgets = [self.description, pb.Percentage(), ' ',   
                   pb.Bar(marker=pb.RotatingMarker()), ' ', pb.ETA()]
        self.timer = pb.ProgressBar(widgets=widgets, maxval=self.n_iter).start()

    def update(self, q=1):
        #update timer
        self.timer.update(self.iter)
        self.iter += q

    def finish(self):
        #end timer
        self.timer.finish()
```

## Function to assign unique values to unique netlist

```python
def to_ID(df,dictionary):
    
    def ID(nets):
        n = []
        if isinstance(nets,str):
            n.append(dictionary[nets])
            #print(n)
        elif isinstance(nets,list):
            n.extend(map(lambda x: dictionary[x],nets))
            
        return n
    df['Net_ID'] = df['Net'].apply(lambda x: ID(x))
    df['Inputs_ID'] = df['Inputs'].apply(lambda x: ID(re.findall("\((.*?)\)",x)))
    return df
```

```python
def dict_func(df):
    l = []
    
    def extract(nets):
        if isinstance(nets,str):
            l.append(nets)
        elif isinstance(nets,list):
            l.extend(nets)
            
    df['Inputs'].apply(lambda x: extract(re.findall("\((.*?)\)",x)))
    df['Net'].apply(lambda x: extract(x))
    
    print("Nets: ",len(l))
    keys = list(set(l))
    print("Unique Nets:",len(keys))
    d = defaultdict(count(1).__next__)
    values = [d[k] for k in keys]
    dictionary = dict(zip(keys, values))
    net_df = pd.DataFrame(keys, index=values, columns =['Net']) 
    net_df['ID'] = values
    
    df = to_ID(df,dictionary)
    net_df.to_csv("net_df", encoding='utf-8', index=False)
    
    
    
    return df, dictionary
```

```python
#Main Function

def ff_out(level,nets,num=0):
    t = []
    #print(num)
    if isinstance(nets,list):
        for i in nets:
            #print("Checking ",i)
            val,temp = match(i)
            if val == 1:
                #print("Got it")
                t.extend(temp)
                t = list(dict.fromkeys(t))
            
    elif isinstance(nets,str):
        #print("Checking ",nets)
        val, temp = match(nets)
        if val == 1:
            t.extend(temp)
        else:
            return 0                     #Return 0 when no instance is found
        
    t = list(filter(None,t)) 
    
    
    if level>1:
        #print("\n\n\n\n")
        level = level-1
        #print("Im here",level+1)
        temp = map(select,t)
        temp = list(filter(None,temp)) 
        a = len(temp)
        return ff_out(level,t,a)          #Recursive function for moving forwards to the given level
        
    #print("Woohoo! I'm going home",level)
    #print("\n\n\n\n\n\n\n")
    #print(t)
    t = map(select,t)
    t = list(filter(None,t)) 
  
    a = len(t)       
    return a+num
```

```python
def out_nearest_flipflop(df):
    
    def select(t):
        temp = df.loc[df["Net"] == t]['Type'].values[0]
        if temp == 'SDFFX1':
            return 1
        else:
            return 0
    
    
    def match_pout(net):      
        net_t = "(" + net + ")"
        temp = new[new['Inputs'].str.contains(net_t, regex=False)]
        
        if temp.empty == False:
            return 1,list(temp['Net'].values)
        else:
            return 0,None 
        
        
    def pout_func(nets,level=1):
        t = []
        flag = 1

        if isinstance(nets,list):
            for i in nets:
                val,temp = match_pout(i)
                if val == 1:
                    t.extend(temp)
                    t = list(set(t))
                else:
                    flag = val

        elif isinstance(nets,str):
            #print("Checking ",nets)
            val, temp = match_pout(nets)
            if val == 1:
                t.extend(temp)
                
            else:
                flag = val
                                            #SET flag = 0 when its a dead end

        if flag==1:
            t = list(filter(None,t)) 
            for i in t:
                buff = select(i)
                if buff:
                    return level
                
        elif flag==0:
            return -1                      #Return -1 when no SDFFX1 is found

        #print("\n\n\n\n")
        level = level+1
        #print("Im here",level+1)
        return pout_func(t,level)
    
    tqdm.pandas()
    return df['Net'].progress_apply(lambda x:pout_func(x))

```

```python
def out_nearest_mux(df):
    
    def select(t):
        temp = df.loc[df["Net"] == t]['Type'].values[0]
        if temp == 'SDFFX1':
            return 1
        else:
            return 0
    
    
    def match_pout(net):      
        net_t = "(" + net + ")"
        temp = new[new['Inputs'].str.contains(net_t, regex=False)]
        
        if temp.empty == False:
            return 1,list(temp['Net'].values)
        else:
            return 0,None 
        
        
    def pout_func(nets,level=1):
        t = []
        flag = 1

        if isinstance(nets,list):
            for i in nets:
                val,temp = match_pout(i)
                if val == 1:
                    t.extend(temp)
                    t = list(set(t))
                else:
                    flag = val

        elif isinstance(nets,str):
            #print("Checking ",nets)
            val, temp = match_pout(nets)
            if val == 1:
                t.extend(temp)
                
            else:
                flag = val
                                            #SET flag = 0 when its a dead end

        if flag==1:
            t = list(filter(None,t)) 
            for i in t:
                buff = select(i)
                if buff:
                    return level
                
        elif flag==0:
            return -1                      #Return -1 when no SDFFX1 is found

        #print("\n\n\n\n")
        level = level+1
        #print("Im here",level+1)
        return pout_func(t,level)
    
    tqdm.pandas()
    return df['Net'].progress_apply(lambda x:pout_func(x))


```
